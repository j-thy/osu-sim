```python
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
```

```python
angles = []

# Count the total number of angle values in angles.txt
num_lines = sum(1 for line in open('angles.txt'))

# Import angles.txt, which contains every angle value across all beatmaps in osu!
with open('angles.txt') as f:
    for line in tqdm(f, total=num_lines):
        angles.append(float(line.strip()))
```

```python
# Create a kernel density estimate plot of the angle values to visualize the most common angles
plot = sns.kdeplot(angles)

# Create labels and title for the plot
plt.xlabel('Angle (degrees)')
plt.ylabel('Density')
plt.title('Cumulative Angle Distributions of All Beatmaps')

# Handpick bucket bounds based on the maxima and minima from the plot and knowledge of osu! gameplay
angle_breaks = [15, 75, 130]

# Shade the bucket regions with different colors
plt.axvspan(0, angle_breaks[0], alpha=0.5, color='red')
plt.axvspan(angle_breaks[0], angle_breaks[1], alpha=0.5, color='green')
plt.axvspan(angle_breaks[1], angle_breaks[2], alpha=0.5, color='blue')
plt.axvspan(angle_breaks[2], 180, alpha=0.5, color='yellow')

# Plot the graph
plt.xlim(0, 180)
plt.show()
```
