```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
import math

# Number of angle buckets
n_a = 10
# Number of distance buckets
n_d = 32
# Size of each angle bucket (180 degrees / 10 buckets = 18 degrees)
s_a = 180 / n_a
# Size of each distance bucket (640 units / 32 buckets = 20 units)
s_d = 640 / n_d

# Create an array of size n_a, initialized to 0
angle_arr = np.zeros(n_a)
# Create an array of size n_d, initialized to 0
dist_arr_not_scaled = np.zeros(n_d)
dist_arr_scaled = np.zeros(n_d)

for a in range(180):
    b_a = math.floor(a / s_a) if a < 180 else n_a - 1
    angle_arr[b_a] += 1

for d in range(640):
    b_d = math.floor(d / s_d) if d < 640 else n_d - 1
    dist_arr_not_scaled[b_d] += 1
    b_d = math.floor((1 - (b_d / (n_d - 1)) ** 2) * (n_d - 1))
    dist_arr_scaled[b_d] += 1

# Plot angle_arr as a bar graph
bar = sns.barplot(x=np.arange(n_a), y=angle_arr)
bar.set(xlabel='Angle Buckets', ylabel='Number of Occurrences', title='Angle Distribution')
plt.show()

bar = sns.barplot(x=np.arange(n_d), y=dist_arr_not_scaled)
bar.set(xlabel='Distance Buckets', ylabel='Number of Occurrences', title='Distance Distribution')
plt.show()

bar = sns.barplot(x=np.arange(n_d), y=dist_arr_scaled)
bar.set(xlabel='Distance Buckets', ylabel='Number of Occurrences', title='Distance Distribution')
plt.show()
```
